***Punto 3:*** Explique como se clasifican los enfoques de aprendizaje automático y defina cada uno de ellos.

Los enfoques de aprendizaje automático se clasifican en cuatro categorías:

- Aprendizaje Supervisado: en este aprendizaje se utiliza un conjunto de datos etiquetados, donde cada ejemplo de entrenamiento se compone de una entrada y una salida- tiene como objetivo aprender una función o modelo que pueda mapear entradas y salidas correctas con mucha precisión.

- Aprendizaje No Supervisado: en este aprendizaje se utilizan datos no etiquetados, datos que no tienen salidas definidas. Como objetivo tiene el descubrir patrones, relaciones o estructuras ocultas de datos.

- Aprendizaje Semi-Supervisado: en este aprendizaje tenemos una especie de fusión de aprendizajes en los cuales se uno aprendizaje supervisado y no supervisado, utiliza un pequeño conjunto de datos etiquetados junto con un conjunto mucho mas grande de datos no etiquetados. Mejora la precisión del modelo con una menor cantidad de datos etiquetados, estos datos suelen ser costosos o difíciles de obtener.

- Aprendizaje por Refuerzo: en este aprendizaje tenemos un agente que interactua con un entorno y aprende a tomar decisiones para maximizar una recompensa acumulada a lo largo del tiempo, como peculiaridad el agente no recibe instrucciones explicitas, sino que aprende a través de prueba y error.

***Punto 10. (1.0)*** Se le proporciona un conjunto de datos denominado `medical_insurance` que contiene información sobre seguros médicos. Este conjunto de datos incluye diversos factores que influyen en los gastos médicos, como la edad, el sexo, el índice de masa corporal (BMI), el hábito de fumar, el número de hijos y la región. Este conjunto de datos se utiliza para entrenar modelos de aprendizaje automático que pueden predecir los gastos médicos de los nuevos asegurados.

Su tarea es la siguiente:

1. Cargue el archivo `medical_insurance` en su entorno de trabajo en Collab.

2. Realice un análisis exploratorio de los datos para entender las características del conjunto de datos.

3. Desarrolle un modelo de regresión lineal que establezca una relación entre el BMI y los gastos que normalmente se dan en un hospital.

4. Una vez que haya entrenado su modelo, utilícelo para predecir los gastos médicos para personas con un BMI de 24.5, 36.3 y 28.8.

```python
# 1. Cargue el archivo `medical_insurance` en su entorno de trabajo en Collab.

import pandas as pd

df = pd.read_csv('Datasets/medical_insurance.csv')

df.head()
```

```python


# 2. Realice un análisis exploratorio de los datos para entender las características del conjunto de datos.

# 3. Desarrolle un modelo de regresión lineal que establezca una relación entre el BMI y los gastos que normalmente se dan en un hospital.

import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['bmi'], y=df['charges'])
plt.title('Relación entre BMI y los Gastos Médicos')
plt.xlabel('BMI')
plt.ylabel('Gastos Médicos')    
plt.show()

# Gráfico de dispersión con línea de regresión
plt.figure(figsize=(10, 5))
sns.regplot(x=df['bmi'], y=df['charges'], line_kws={"color":"red"})
plt.title('Relación entre BMI y los Gastos Médicos con Línea de Regresión')
plt.xlabel('BMI')
plt.ylabel('Gastos Médicos')
plt.show()




```

```python

#4. Una vez que haya entrenado su modelo, utilícelo para predecir los gastos médicos para personas con un BMI de 24.5, 36.3 y 28.8.


from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

X = df[['bmi']]  
y = df['charges'] 


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)


model = LinearRegression()


model.fit(X_train, y_train)


y_pred = model.predict(X_test)

bmi_values = [[24.5], [36.3], [28.8]]


predictions = model.predict(bmi_values)


for bmi, charge in zip([24.5, 36.3, 28.8], predictions):
    print(f'Predicción de gastos médicos para BMI {bmi}: {charge}')
```

***Punto 11 (1.0)*** Desarrolle un algoritmo de programación para un cajero automático que tendrá un saldo inicial de un millón de pesos, el cajero debe tener un menú con las siguientes opciones 

• Retirar dinero de la cuenta (El usuario debe ingresar el valor a retirar, y se debe imprimir un mensaje del saldo restante, tenga en cuenta que el usuario podría digitar un valor de salida más grande de lo que hay en el saldo, por lo tanto, el cajero se lo puede hacer saber al usuario mediante un mensaje de error).

• Ingresar dinero en la cuenta (El usuario debe ingresar el valor a digitar, y se le debe imprimir el saldo actual después del ingreso).}

• Mostrar saldo (imprimir el saldo actual).

• Salir (Imprimir un mensaje de agradecimiento por utilizar el cajero).

```python
print("### Bienvenido a su Cajero Automatico ###")

saldo = 1000000
din_actual = 0

while True:
  opcion = int(input("Opciones Disponibles: \n1.Retira Dinero \n2.Ingresa Dinero \n3.Saldo Disponible \n4.Salir \nSeleccione una de las siguientes opciones: "))

  if opcion == 1:
    dinero = int(input("Seleccione la cantidad de dinero que desea retirar: "))
    if din_actual == 0:
      din_actual = saldo
    else:
      din_actual = din_actual

    if dinero <= din_actual:
      din_actual = din_actual - dinero
      print("Dinero actual en el cajero: $",din_actual)
    else: 
      print("Cantidad de dinero no disponible")

  elif opcion == 2:
    dinero = int(input("Seleccione la cantidad de dinero que desea ingresar: "))
    if din_actual == 0:
      din_actual = saldo + dinero
    else:
      din_actual = din_actual + dinero
    print("Dinero actual en el cajero: $",din_actual)

  elif opcion == 3:
    if din_actual == 0:
      din_actual = saldo
    else:
      din_actual = din_actual

    print("Saldo actual: $", din_actual)

  elif opcion == 4:
    print("Gracias por utilizar nuestros cajeros de confianza.")
    break

  else:
    print("Opcion no valida\n")
```

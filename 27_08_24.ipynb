<p>Media: El valor promedio</p>
<p>Mediana: El valor medio</p>
<p>Moda: El valor mas repetido</p>

<p>Desviacion Estandar </p>

```python
import numpy as np
from scipy import stats
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]

x = np.mean(speed)
y = np.median(speed)
z = stats.mode(speed)

print("Media: ", x)

print("Mediana: ", y)

print("Moda: ", z.mode, "or", z)
```

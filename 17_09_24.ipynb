<h3>Regrension Polimonica:</h3> Si sus puntos de datos claramente no se ajustan a una regresion lineal, podria ser ideal para la regresion polinómica.

<p>Formula: y = b0 + b1.x1 + b2.x1^2 + ... + bn.x1^n</p>

<p>--> Aproximadamente todos los datos del mundo real son de naturaleza no lineal</p>


<h3>r^2 (Coeficiente de determinación)</h3>

<p>Es un valor que oscila entre 0 y 1, donde 0 significa que nor hay relación, y 1 significa 100% relación</p>

<h3>Ejemplo:</h3>
<p>Hemos registrado 18 coches al pasar por un peaje determinado. Registramos la velocidad del coche y el momento del dia(hora) en que se produjo el paso.</p>

<p>x : Horas del dia</p>
<p>y : Velocidad del automovil</p>

```python
import matplotlib.pyplot as plt


x = [1, 2, 3, 5, 6, 7, 8, 9, 10, 12, 13, 14, 15, 16, 18, 19, 21, 22]
y = [100, 90, 80, 60, 60, 55, 60, 65, 70, 70, 75, 76, 78, 79, 90, 99, 99, 100]


plt.scatter(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy

x = [1, 2, 3, 5, 6, 7, 8, 9, 10, 12, 13, 14, 15, 16, 18, 19, 21, 22]
y = [100, 90, 80, 60, 60, 55, 60, 65, 70, 70, 75, 76, 78, 79, 90, 99, 99, 100]

mymodel = numpy.poly1d(numpy.polyfit(x, y, 13)) #Ajusta un polimonio cubico (grado 3) a los datos y crear un modelo.

myline = numpy.linspace(1, 22, 100) # Crea un arreglo con 100 valores espaciados uniformemete de 1 a 22

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()
```

```python
import numpy
from sklearn.metrics import r2_score

x = [1, 2, 3, 5, 6, 7, 8, 9, 10, 12, 13, 14, 15, 16, 18, 19, 21, 22]
y = [100, 90, 80, 60, 60, 55, 60, 65, 70, 70, 75, 76, 78, 79, 90, 99, 99, 100]

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

print(r2_score(y, mymodel(x)))
```

<h1>Regresión Multiple</h1>

<p>Es como la regresión lineal, pero con más de un valor independiente, lo que significa que intentamos predecir un valor basandonos en dos o más variables.</p>

```python
# 1. Cargue el archivo `medical_insurance` en su entorno de trabajo en Collab.

import pandas as pd

df = pd.read_csv('Datasets/Cars.csv')

df.head()
```

```python
import matplotlib.pyplot as plt
from sklearn import linear_model
X = df[['Weight', 'Volume']]
y = df['CO2']

regr = linear_model.LinearRegression()
regr.fit(X, y)

predictedCO2 = regr.predict([[1200, 1160]])

print(predictedCO2)
```
